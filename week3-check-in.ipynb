```python
import numpy as np
import pandas as pd

# Import data
URL = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/main/inst/extdata/penguins.csv'
penguins = pd.read_csv(URL)
```

```python
(penguins.groupby(['year', 'species'])
         .body_mass_g
         .max()
         .rename('max_body_mass_g')
)
```

```python
# if you put it in () then it will chain everything
(penguins.groupby(['year','species'])
         .body_mass_g
         .max()
         .sort_values()
         .plot(kind ='barh',
               title ='Penguin mass body mass by species and year',
               xlabel =('Year', 'Species'))
         .set_xlabel(xlabel='Body mass(g)')
)
```
